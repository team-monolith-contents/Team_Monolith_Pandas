> 위 그림은 Panel data 전형적인 예시를 보여줍니다. 이처럼 행열의 2차원 구조를 가진 데이터를 일반적으로 Panel Data라고 부릅니다. </br></br>
이러한 데이터는 항상 "행"과 "열"로 데이터를 구분합니다. 특히 위 그림처럼 숫자 뿐 아니라 문자도 포함할 수 있습니다. 이러한 데이터가 행과 열로 구조화되어있다고 하여 구조화된 데이터(Structured Data)라고도 부릅니다. </br></br>많은 데이터가 이러한 형태를 지니고 있기 때문에 파이썬에서는 이를 위한 전용 패키지를 만들었습니다. 이것이 바로 Pandas입니다.

## **판다스 데이터의 구조**

- 판다스의 데이터는 List, Numpy array와는 다르게 항상 Key, Value의 구조를 가지고 있습니다. 그런데 Key, Value라는 표현은 앞선 데이터의 종류 중에서 딕셔너리(Dictionary) 데이터에서도 나오는 개념입니다.

- 판다스도 key와 value의 구조를 지니며 이 딕셔너리 데이터 구조와 매우 유사합니다.

파이썬의 numpy 패키지와 같이 pandas또한 패키지로 구현되어 있습니다. 

#### 따라서 Pandas를 활용하기 위해서는 *import pandas as pd* 로 불러옵니다. 일반적으로 pandas는 pd로 축약하여 활용합니다.

```python
import pandas as pd # 판다스를 pd로 축약하여 불러옵니다
```

### **2. Pandas 데이터 정의해보기**
> 위에서 설명하였듯, 행과 열로 2차원 구조를 가지는 판다스 데이터는 항상 key, value를 가지고 있습니다.</br>
따라서 먼저 가장 기본이 되는 key, value를 가지는 딕셔너리 데이터를 정의해보겠습니다.

```python
# 예제 1: 파이썬 dictionary를 활용하여 key, value로 데이터 구성하기

items = {'apple':1000, 
         'cup': 2000, 
         'pen': 500, 
         'banana': 4000}
print("생성된 딕셔너리: ", items)

# key로 특정 value를 불러오기
print("apple의 value: ", items["apple"])
print("pen의 value: ", items["pen"])
```

### **3. 판다스 데이터 프레임의 구조**

> 위에서 만들어진 데이터를 살펴보면 딕셔너리의 key가 열로, 각 value가 행으로 전환된 것을 확인할 수 있습니다. 특히 숫자 뿐만 아니라 문자열 데이터(price, pcs)도 잘 표현되어있는 것을 볼 수 있습니다.

- 판다스는 위 그림처럼 기본적으로 Series가 모여서 하나의 DataFrame을 이룹니다. 위에서 간단한 딕셔너리 데이터 예시로 생성한 것이 바로 DataFrame인 것이죠. 

- Series: 전체 데이터에서 1개 열의 데이터 </br>
> 활용 방법 : pd.Series(데이터 변수)

- DataFrame: 여러 Series가 합쳐진 구조 </br>
> 활용 방법 : pd.DataFrame(데이터 변수)

```python
# 예제 4 : 판다스 es 데이터 생성해보기

# 판다스 Series 데이터 생성 
ser = pd.Series(['a','b','c',3])  # index값과 함께 출력됩니다. 
print("판다스 Series:\n", ser)

# 국가번호 예시
Nations_Numbers = {'Korea': 82, 'America': 1, 'Swiss': 41, 'Italy': 39, 'Japan': 81, 'China': 86, 'Rusia': 7}
ser = pd.Series(Nations_Numbers)
print("\n판다스 Series:\n", ser)
```

### <연습문제 Set>
<a href='#jce-judge/open/A2ZS8OQTZmE' target='_self'><h3>판다스 기초 4 - 시리즈</h3></a>

- 판다스 Series는 데이터 이름 등의 속성을 설정할 수 있습니다. 

- series 데이터 이름 정하기 </br>
> 활용 방법 : 데이터 변수.name

- index의 이름 정하기 </br>
> 활용 방법 : 데이터 변수.index.name

```python
# 예제 5 : 판다스 Series 데이터에 "이름"과 "index의 이름"을 설정해보기

# 판다스 Series 데이터 생성 
ser = pd.Series(['a','b','c',3]) 
print("판다스 Series:\n", ser)

# 현재는 None이 출력
print(ser.name) 
print(ser.index.name)

# 이름 설정
ser.name = '국가 번호'
ser.index.name = '국가명'

print("\n판다스 Series:\n", ser) # Series데이터의 이름과 index의 이름이 설정된 것을 볼 수 있습니다.

# 예제 5-1
# 새로운 판다스 Series 데이터를 생성하고, 데이터 이름 설정하기
ser = pd.Series({'health': 80, 'living': 40, 'food': 100}) 
ser.name = '중요도'
ser.index.name = '삶의 요소'

print("\n판다스 Series:\n", ser)
```

### **4. 하나의 key에 여러 value를 가지는 딕셔너리를 활용하여 Series와 DataFrame 데이터 만들기**

> 1. Dictionary를 생성한다. 단, 이때에는 하나의 key에 여러개의 value를 넣어주세요. </br>
2. 바로 위에서 다루었던 pd.Series나 pd.DataFrame을 활용하여 판다스 데이터를 만듭니다.

```python
# 예제 6 : 하나의 key에 여러 value를 가지는 딕셔너리에 Series와 DataFrame을 적용해보자
data = {'Region' : ['Korea', 'America', 'Chaina', 'Canada', 'Italy'],
        'Sales' : [300, 200, 500, 150, 50],
        'Amount' : [90, 80, 100, 30, 10],
        }

# Series 적용
ser = pd.Series(data)
ser
```

```python
# DataFrame 적용
df = pd.DataFrame(data)
df
```

#### **여기서 잠깐!**
> 위의 결과를 보면 Series와 DataFrame의 차이를 확인할 수 있습니다. </br>
`Series는 항상 1개의 열`로 만들어지는 반면, `DataFrame은 여러개의 열`을 가질 수 있습니다.

#### 또한 판다스 DataFrame도 Series와 마찬가지로 각각의 행과 열에 원하는 이름을 설정할 수 있습니다.

```python
# 예제 7 : 판다스 DataFrame의 각 행과 열에 이름 붙이기

# 앞선 예제와 동일한 데이터프레임 생성하기
data = {'Region' : ['Korea', 'America', 'Chaina', 'Canada', 'Italy'],
        'Sales' : [300, 200, 500, 150, 50],
        'Amount' : [90, 80, 100, 30, 10],
        }
df = pd.DataFrame(data)

df.index=['one','two','three','four','five'] # 각 행의 이름을 변경합니다. . 
df.columns = ['국가명','가격','수량'] # 각 열의 이름을 변경합니다. 
df
```
